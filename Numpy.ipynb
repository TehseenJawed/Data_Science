```python
import numpy as np
import matplotlib.pylab as plt

def plti(im, h=8, **kwargs):
    """
    Helper Function to plot an image
     """
    y=im.shape[0]
    x=im.shape[1]
    w=(y/x)*h
    plt.figure(figsize=(w,h))
    plt.imshow(im,interpolation="none", **kwargs)
    #plt.axis('off')
%matplotlib inline
im= plt.imread("TJ CV.jpg")
#im = im * 0
print(im.shape, type(im))
#plti(im)
im=im[:,:,:]
plti(im)
#print(im.shape)
   
```

# Split

```python
arr= np.array([[1,2,3,4], [1,2,3,4]])
arr1= np.array([[2,3,4,5],[2,3,4,5]])
na3 = np.concatenate([arr,arr1], axis = 1) # concatinate by column through axis 1
print(na3)
print(np.split(na3,[2,5,8]))
```
